```python
#Sonuçların Görselleştirilmesi (AI CODE)
plt.scatter(X, y, color="blue", label="Gerçek Veri")  # Gerçek veri noktaları
plt.plot(X, y_sapka, color="red", label="Model Tahmini")  # Modelin tahmin ettiği doğru
plt.xlabel("Bağımsız Değişken (X)")
plt.ylabel("Bağımlı Değişken (y)")
plt.legend()
plt.title("Linear Regresyon (Least Squares Estimation)")
plt.show()
```
